This script gives a visualization of the decision tree of the inputted dataset.

```python
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# import dataset
path = 'https://raw.githubusercontent.com/scraus/cmsc173-project1/main/movie.csv'
dataset = pd.read_csv(path)
dataset = np.array(dataset);
dataset
```

```python
# extract ratings column and change positive values to 1 and negative values to 0
ratings = dataset[:, 0]
ratings[ratings >= 0] = 1
ratings[ratings < 0] = 0
ratings
```

```python
# display other features / independent variables
examples = dataset[:, 1:]
examples
```

```python
# define column headers
features = ['over2hours','sci-fi','action','drama','awardee']
```

```python
# initialize decision tree
decision_tree = tree.DecisionTreeClassifier()
decision_tree = decision_tree.fit(examples, ratings)

# decision tree size rendering configurations
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(5)

# decision tree visualization
tree.plot_tree(decision_tree, feature_names=features, filled=True)
```
